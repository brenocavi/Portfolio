# Panorama do COVID-19 

O conovavírus (COVID-19) é uma doenção causada pelo vírus SARS-Cov-2.

Os meios de transmissão se dão por gotículos espelidas, quando pessoas infectadas espirram, tossem, falam ou até mesmo respiram. O grau de intensidade dos sintomas variam entre pessoas, podendo ser uma falta de ar leve até casos de intervação em UTI, e alguns casos morte. Existe pessoas que são infectadas, porém, não desenvolve nenhum sintoma. 



<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Pessoas que tiveram grau mais elevado dos sintomas, relataram a perda temporário de olfato e paladar, além de ter capacidade pulmonar comprometida. 

Os dois últimos anos, houveram diversas medidas adotadas, a fim de, tentar combater o avanço o vírus. O isolamento da população foi o meio preventivo adotado, até disponibilização de vacinas, que até então não existiam. 

Para explicação mais completa sobre o vírus: [Organização Mundial da Saúde]( https://www.who.int/news-room/questions-and-answers/item/coronavirus-disease-covid-19-how-is-it-transmitted)

A análise presente aqui, visa demonstram a evolução dos quadros de contaminção, obitos e vacinação em escala mundial, bem como a situação do Brasil.Foram ultilizados dados públicos da doença.

Obs: *O trabalho aqui presente, não possui viés politico e  nem visa descorrer sobre medidas adotadas por nenhum país e seus governantes*

## Obtenção dos Dados

Os dados trabalhados aqui são disponibilizados pela Global Change Data Lab, uma instituição sem fins lucrativos que tem por proposito compartilhar dados da condições de vida global e das mudanças ambientais. A abordagem para os dados da COVID-19 podem ser encontrados no [Our World in Data](https://ourworldindata.org/coronavirus).

A coleta dos dados é feita a partir da união de diferentes instituições, como Johns Hopkins, o próprio Our World in Data e Governos Federativos. O resultado dessa junção de dados, resultou em um Dataset que sofreu algumas alterações, desde do ínicio dos registros, tendo aumentado o tamanho de varíaveis. Com essa constante alteração, é possível que exista uma grande quantidade de dados faltantes, uma vez, que novas varíaveis antes não eram observadas ou registradas. 


O registro dos dados são atualizados diariamente, e continuam a ser registrados.


### Dicionário de Variáveis
* **iso_code**: Código de três letras definido pela ISO 3166-1 alpha-3 (padronização de países, territorios dependentes, zonas especiais de interesse geográfico);
* **continent**: Continente da localização geográfica;
* **location**: Países e consideráveis regiões especias;
* **date**: Dia;
* **total_cases**: Total de casos confirmados;
* **new_cases**: Novos casos confirmados;
* **new_cases_smoothed**: Novos casos confirmados, suaviados exponencialmente pelo últimos 7 dias;
* **total_deaths**: Total de obitos;
* **new_deaths**: Número de mortes confirmados pelo vírus;
* **new_deaths_smoothed**: Número de mortes confirmados pelo vírus, suaviados exponencialmente pelo últimos 7 dias;
* **total_cases_per_million**: Total de casos confirmados por milhão de pessoas;
* **new_cases_per_million**: Total de novos casos confirmados por milhão de pessoas;
* **new_cases_smoothed_per_million**: Total de novos casos confirmados por milhão de pessoas, suaviados exponencialmente pelo últimos 7 dias;
* **total_deaths_per_million**: Total de mortes confirmadas pelo vírus por milhão de pessoas;
* **new_deaths_per_million**: Total de novas mortes confirmadas pelo vírus por milhão de pessoas;
* **new_deaths_smoothed_per_million**: Total de novas mortes confirmadas pelo vírus por milhão de pessoas, suaviados exponencialmente pelo últimos 7 dias;
* **reproduction_rate**: Taxa estimada de reprodução do vírus;
* **icu_patients**: Número de pacientes internados em Unidads de Tratamento Intensivo;
* **icu_patients_per_million**: Número de pacientes internados em Unidads de Tratamento Intensivo por milhão de pessoas;
* **hosp_patients**: Número de pessoas internadas por conta do vírus;
* **hosp_patients_per_million**: Número de pessoas internadas por conta do vírus por milhão de pessoas;
* **weekly_icu_admissions**: Número de novas pessoas internados em Unidade de Tratamento Intensivo por semana;
* **weekly_icu_admissions_per_million**: Número de novas pessoas internados em Unidade de Tratamento Intensivo por semana milhão de pessoas;
* **weekly_hosp_admissions**: Número de novas pessoas internadas por conta do vírus;
* **weekly_hosp_admissions_per_million**: Número de novas pessoas internadas por conta do vírus por milhão;
* **total_tests**: Total de testes realizados;
* **new_tests**: Número de novos testes realizados;
* **total_tests_per_thousand**: Total de testes realizados a cada mil pessoas;
* **new_tests_per_thousand**: Número de novos testes realizados a cada mil pessoas;
* **new_tests_smoothed**: Número de novos testes realizados, suaviados exponencialmente pelo últimos 7 dias;
* **new_tests_smoothed_per_thousand**: Número de novos testes realizados a cada mil pessoas, suaviados exponencialmente pelo últimos 7 dias;
* **positive_rate**: Média de testes com resultado positivo;
* **tests_per_case**: Média de testes realização em relação a novos casos confirmados;
* **tests_units**: Tipo de unidade utilizada, por cada país, para relatar o resultado obtidos pelos teste;
* **total_vaccinations**: Quantidade total de vacinas aplicadas;
* **people_vaccinated**: Quantidade de pessoas que receberam pelo menos uma dose da vacina;
* **people_fully_vaccinated**: Quantidade de pessoas com ciclo vacinal completo;
* **total_boosters**: Quantidade de doses de reforço aplicadas;
* **new_vaccinations**: Quantidade de novas vacinas aplicadas;
* **new_vaccinations_smoothed**: Quantidade de novas vacinas aplicadas, suaviados exponencialmente pelo últimos 7 dias;
* **total_vaccinations_per_hundred**: Quantidade total de vacinas aplicadas a cada cem pessoas;
* **people_vaccinated_per_hundred**: Quantidade de pessoas que receberam pelo menos uma dose da vacina a cada cem pessoas;
* **people_fully_vaccinated_per_hundred**: Quantidade de pessoas com ciclo vacinal completo a cada cem pessoas;
* **total_boosters_per_hundred**:  Quantidade de doses de reforço aplicadas a cada cem pessoas;
* **new_vaccinations_smoothed_per_million**:  Quantidade de novas vacinas aplicadas a cada milhão de pessoas;
* **new_people_vaccinated_smoothed**: Quantidade de novas pessoas vaciandas, suavizados exponencialmente pelos últimos 7 dias;
* **new_people_vaccinated_smoothed_per_hundred** Quantidade de novas pessoas vaciandas, suavizados exponencialmente pelos últimos 7 dias a cada cem pessoas;
* **stringency_index**: Índice de Rigidez de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibição de viagens, reescalonado para um valor de 0 a 100;
* **population**: Habitantes do país (dados de 2019);
* **population_density**: Densidade populacional;
* **median_age**: Idade média da população;
* **aged_65_older**: Parcela da população com 65 anos ou mais, ano mais recente disponível;
* **aged_70_older**: Parcela da população com 70 anos ou mais (dados de 2015);
* **gdp_per_capita**: Produto interno bruto em paridade de poder de compra (dólares internacionais constantes de 2011), ano mais recente disponível;
* **extreme_poverty**: Parcela da popula que viva em condição de pobreza extrema, ano mais recente disponível;
* **cardiovasc_death_rate**: Taxa de mortalidade por doença cardiovascular em 2017 (número anual de mortes por 100.000 pessoas);
* **diabetes_prevalence**: Parcela da população entre 20 a 79 anos, que possui diabetes;
* **female_smokers**: Parcela da população feminina fumante, ano mais recente disponível;
* **male_smokers**: Parcela da população masculina fumante, ano mais recente disponível;
* **handwashing_facilities**: Percentagem da população com saneamento básico, ano mais recente disponível;
* **hospital_beds_per_thousand**: Leitos hospitalares a cada milhão de pessoa;
* **life_expectancy**: Expectativa de vida;
* **human_development_index**: Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano – uma vida longa e saudável, conhecimento e um padrão de vida decente;
* **excess_mortality**: diverça de mortes registradas entre 2020 e 2021, para o mesmo período do ano.
* **excess_mortality_cumulative**: Diferença percentual entre o número acumulado de mortes desde 1 de janeiro de 2020 e as mortes cumulativas projetadas para o mesmo período com base em anos anteriores;
* **excess_mortality_cumulative_absolute**: Diferença absoluta entre o número acumulado de mortes desde 1 de janeiro de 2020 e as mortes cumulativas projetadas para o mesmo período com base em anos anteriores;
* **excess_mortality_cumulative_per_million**: Diferença absoluta entre o número acumulado de mortes desde 1 de janeiro de 2020 e as mortes cumulativas projetadas para o mesmo período com base em anos anteriores, a cada milhão de pessoas.

###Evolução Mundial

Com uma ideia inicial do que o está presente no Dataset, é interessante seguir para uma analise do contexto da evolução do virus quanto a total de casos, total de obitos, novos casos diarios, pessoas vacinadas.

Dentro coluna *location* , designada para alocar os países, existe um elemente denominado WORLD. Ao realizar um filtro com esse elemento, é possível ter um Dataset apenas com dados mundiais.

```
#Filtro para dados mundiais
world = df.loc[df.location == 'World']
#Transformar a coluna date em index
world.set_index('date', inplace=True)
```

```
#gráfico de evolução mundial 
fig, ax = plt.subplots(nrows= 2 , ncols= 1, figsize=(10,8))

#evolução de novos casos
#world['new_cases'].plot( ax=ax[0])
world['new_cases_smoothed'].plot( ax=ax[0])
ax[0].set_ylabel('Novos Casos')
#ax[0].set_title('Evolução Mundial de Novo Casos')

#evoluação do total de casos
world['total_cases'].plot(ax=ax[1])
ax[1].set_ylabel('Total de Casos')
#ax[1].set_title('Evolução Mundial do Total de Casos')

fig.tight_layout()
```

```
world.loc[world.total_cases != 0.0].head(1)
```

```
#Verificado primeiro dia de obito
world.loc[world.total_deaths != 0.0].head(1)
```

Após cerca de **22 dias**  do ínicio dos registros do Dataset, foram incluidos os primeiros casos e os primeiros obitos. Cabe salientar que a ausência de dados nos primeiros 21 dias, provavelmente deve-se a falta de parâmetros para identificação da doença e confirmação de sintomas, questões essas que extrapolam as análises aqui descritas. 

```
#gráfico de evolução mundial 
fig, ax = plt.subplots(nrows= 2 , ncols= 1, figsize=(10,8))

#evolução de novos casos
world['new_deaths_smoothed'].plot( ax=ax[0])
ax[0].set_ylabel('Novas Mortes')
ax[0].set_title('Novas Mortes Diarias')

#evoluação do total de casos
world['total_deaths'].plot(ax=ax[1])
ax[1].set_ylabel('Total de Mortes')
ax[1].set_title('Total de Mortes até o Momento')

fig.tight_layout()
```

É possível observar nos gráficos de Novas Mortes Diarias e Evolução Mundial de Casos, que após pico em Janeiro/2022 ocorre queda constante, apesar de apresentar alguns picos menores, o contexto geral é de queda. A data desse pico, em Janeiro, coincide com o que foi classificado como a terceira onda do vírus. 

```
#Organização do plot para 3 gráficos
fig= plt.figure(figsize=(10,8))
gs= gridspec.GridSpec(2,2)

#Gráfico 1
ax = fig.add_subplot(gs[0, :])
ax.plot(world.total_vaccinations)
ax.set_ylabel('Total de Vacinas')
ax.set_xlabel('Data')

#Gráfico 2
ax=fig.add_subplot(gs[1,0])
ax.plot(world.new_vaccinations_smoothed)
ax.set_ylabel('Novas Vacinas')
ax.set_xlabel("Data")

#Gráfico 3
ax=fig.add_subplot(gs[1,1])
ax.plot(world.people_fully_vaccinated)
ax.set_ylabel('Pessoas com Vacinação Completa')
ax.set_xlabel("Data")

fig.tight_layout()
```

```
#Pico de novas vacinas aplicadas
world.loc[world.new_vaccinations == world.new_vaccinations.max() ]
```

Com os gráficos de vacinações é possivel observar o pico de novas vacinas aplicadas foi em Agosto/2021. Confirmado que os valores ausentes da coluna *continent* estão atrelados a divisões já citados, é necessário retirar essas linhas para que assim retorne apenas os países.

```
#Fazer uma cópia do Dataset
df_clean = df.copy()

#Limpando o Dataset apartir a coluna continent
df_clean.dropna(axis=0, subset= ['continent'], inplace=True)

df_cont = df_clean.loc[df_clean.index == '2022-01-19'].sort_values('total_cases', ascending=False)[:10]

df_cont
```

Com Dataset apenas com os 10 países com maior números de casos, fica possível plotar em gráfico para melhor visualização do cenário. 

```
#Gráfico dos 10 paises
fig,ax = plt.subplots(nrows = 2 , ncols= 1, figsize=(10,8))

df_cont.plot(kind='barh', x='location', y='total_cases', ax=ax[0])
ax[0].set_ylabel('Total de Casos')
ax[0].set_title('Total de Casos (19-01-2022)')

#Gáfico ordenado por novos casos
df_cont.sort_values('new_cases', ascending=False).plot(kind='barh', x='location',
                                                       y='new_cases', ax=ax[1])
ax[1].set_ylabel('Novos Casos')
ax[1].set_title('Novos Casos por Pais (19-01-2022)')

fig.tight_layout()
```

Com os gráficos acima vemos os paises com maior número de casos, na data onde foi registrado o recorde mundial de casos em um único dia. Nesta data Os Estados Unidos era o pais com maior quantidade de casos totais, além de ser o pais que apresentou maior número de novos casos. Vale lembrar, que Os Estados Unidos foi amplamente criticado pelas medidas tardias adotadas para combater o avanço do vírus no país. 

Comparando a ordem dos dois gráficos, há um indicação de possível mudança na colocação dos países em número de casos totais, já que os novos casos tiveram registro superiores, como caso da França, de países que estavam em patamares maiores em relação ao número de casos totais.

No cenário atual, vimos que países como França, India e Alemanha subiram no rank, enquanto que o Brasil desceu algumas posições. No próximo gráfico, sobre total de vacinas aplicadas até o momento, é possível ver a situação dos países presentes no gráfico acima.
